```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path, PurePath
from IPython.display import display, HTML
```

```python
amazon_path = "src/main/resources/data/eda-results"
```

## Q1: What is the review count distribution among the star_rating scale for the entire dataset?

```python
star_rating_dist_path = PurePath(amazon_path, "star-rating-dist")
star_rating_df = pd.read_parquet(star_rating_dist_path)
star_rating_df["star_rating"] = star_rating_df["star_rating"].astype(int)
sns.countplot(x="star_rating", data=star_rating_df)
```

## Q2: What is the null count per column of the dataset?

```python
null_path = PurePath(amazon_path, "null-count")
null_count_df = pd.read_parquet(null_path)
df_null_count_unpivoted = null_count_df.melt(var_name="column", value_name="null_count")
sns.barplot(x="column", y="null_count", data=df_null_count_unpivoted)
plt.xticks(rotation = 90)
plt.show()
```

## Q3: What is the average rating per product category?

```python
avg_product_rating_path = PurePath(amazon_path, "avg-product-rating")
avg_product_rating_df = pd.read_parquet(avg_product_rating_path)
g1 = sns.catplot(x = "product_category",
            y = "avg_stars",
            data = avg_product_rating_df,
            kind = "bar",
            errorbar=("ci", True))

plt.xticks(rotation = 75)
plt.show()
```

## Q4: Generate and analyze general statistics for relevante numeric columns

```python
describe_numeric_path = PurePath(amazon_path, "describe")
describe_numeric_df = pd.read_parquet(describe_numeric_path)
display(HTML(describe_numeric_df.to_html()))
```

## Q5: Generate and analyze the word count distribution among all the reviews of the dataset

```python
word_count_path = PurePath(amazon_path, "word-count")
word_count_df = pd.read_parquet(word_count_path)
sns.relplot(x="word_count", y="review_count", data=word_count_df, kind="scatter")
```

## Q5: How would you detect products that should be dropped or kept available?

```python
product_performance_path = PurePath(amazon_path, "product-performance")
product_performance_df = pd.read_parquet(product_performance_path)
product_performance_df["star_rating"] = product_performance_df["star_rating"].astype(int)
sns.pointplot(x="product_id", y="star_rating", data=product_performance_df)
plt.xticks(rotation = 90)
plt.show()
```

## Q7: What is the evolution by year and for every product category of the quantity of reviews?

```python
review_evolution_path = PurePath(amazon_path, "review-qty-evolution")
review_evolution_df = pd.read_parquet(review_evolution_path)
review_evolution_df["year"] = review_evolution_df["year"].astype(int)
sns.relplot(x="year", y="review_count", data=review_evolution_df, kind="line", hue="product_category")
```

## Q8: Is there a bias between verified-purchased reviews and those that are not?

```python
verified_bias_path = PurePath(amazon_path, "verified-bias")
verified_bias_df = pd.read_parquet(verified_bias_path)
sns.barplot(x="product_category", y="avg_rating", hue="verified_purchase", data=verified_bias_df)
plt.xticks(rotation = 90)
plt.show()
```
